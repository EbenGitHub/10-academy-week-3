```python
# Import packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
data_path = '../data/MachineLearningRating_v3.txt'
df = pd.read_csv(data_path, delimiter='|')
```

```python
df.describe()
```

```python
df.tail()
```

```python
print(df.isnull().sum())
print(f"Shape of DataFrame is: {df.shape}")
```

```python
# Drop those with >90% missing data
df = df.drop(columns=['NumberOfVehiclesInFleet', 'CrossBorder'])

# Fill missing values for categorical columns with mode
categorical_columns = ['Bank', 'AccountType', 'MaritalStatus', 'Gender', 'VehicleType', 'make', 'Model', 'bodytype']
for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# Fill missing values for numerical columns with mean or median
numerical_cols = ['CustomValueEstimate', 'mmcode', 'Cylinders', 'cubiccapacity', 'kilowatts', 'NumberOfDoors']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

### Univariate Analysis

```python
# TotalPremium

sns.histplot(df['TotalPremium'], kde=True)
plt.title("Distribution of TotalPremium")
plt.show()
```

```python
# Gender

df['Gender'].value_counts().plot(kind='bar', title='Gender Distribution')
plt.show()
```

```python
# Correlation matrix

corr_matrix = df[['TotalPremium', 'TotalClaims']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Scatter plot of TotalPremium vs. TotalClaims

sns.scatterplot(x='TotalPremium', y='TotalClaims', data=df)
plt.title("TotalPremium vs TotalClaims")
plt.show()
```

```python
# Compare TotalPremium by Province

df.groupby('Province')['TotalPremium'].mean().plot(kind='bar', title='Average Premium by Province')
plt.show()
```

```python
# Outliner

sns.boxplot(df['TotalPremium'])
plt.title("Boxplot of TotalPremium")
plt.show()
```

### Visualization

```python
# Premium Trend Over Time

df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
df.groupby('TransactionMonth')['TotalPremium'].mean().plot(title='Premium Trend Over Time')
plt.show()
```

```python
# Vehicle Make Distribution

df['make'].value_counts().plot(kind='bar', title='Vehicle Make Distribution')
plt.show()
```

```python
# Claims by Zip Code

sns.boxplot(x='PostalCode', y='TotalClaims', data=df)
plt.title("Claims by Zip Code")
plt.show()
```
